```python
b = {}
for w in words:
  chs = ['<S>'] + list(w) + ['<E>']
  for ch1, ch2 in zip(chs, chs[1:]):
    bigram = (ch1, ch2)
    b[bigram] = b.get(bigram, 0) + 1
    # print(ch1, ch2)
```

```python
sorted(b.items(), key = lambda kv: -kv[1])
```

```python
import torch
```

```python
N = torch.zeros((28, 28), dtype=torch.int32)
```

```python
chars = sorted(list(set(''.join(words)))) #중복x
stoi = {s:i for i,s in enumerate(chars)}
stoi['<S>'] = 26
stoi['<E>'] = 27
```

```python
stoi
```

```python
for w in words:
  chs = ['<S>'] + list(w) + ['<E>']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    N[ix1, ix2] += 1
```

```python
N
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(N)
plt.show()
```
